```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
evo_df = pd.read_csv('genetic_evolutions.csv', index_col=0)
```

```python
plt.figure(figsize=(8, 6))

difficulties = ['easy', 'medium', 'hard']

for diff in difficulties:
    means = evo_df.groupby('Difficulty').mean().loc[diff][1:]
    mins = evo_df.groupby('Difficulty').min().loc[diff][1:]
    maxs = evo_df.groupby('Difficulty').max().loc[diff][1:]
    plt.plot(means.index.astype(int), means.values, label=diff)
    plt.fill_between(means.index.astype(int), mins.values, maxs.values, alpha=0.2)

plt.title('Number of Conflicts in Best Solution over Generations')
plt.xlabel('Generations')
plt.ylabel('Minimum Number of Conflicts across Population')
plt.legend()
#plt.savefig('genetic_algorithm_500_gens.png')
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

difficulties = ['easy', 'medium', 'hard']
colors = ['green', 'gold', 'red']

for i in range(len(difficulties)):
    diff = difficulties[i]
    means = evo_df.groupby('Difficulty').mean().loc[diff][1:]
    mins = evo_df.groupby('Difficulty').min().loc[diff][1:]
    maxs = evo_df.groupby('Difficulty').max().loc[diff][1:]
    ax[i].plot(means.index.astype(int), means.values, color=colors[i], label='Mean')
    ax[i].fill_between(means.index.astype(int), mins.values, maxs.values, alpha=0.2, color=colors[i], label='Min/Max')
    ax[i].legend()
    ax[i].set(title=diff.title(), ylim=(0, 40))

plt.suptitle('Number of Conflicts in Best Solution over Generations')
ax[1].set_xlabel('Generations')
ax[0].set_ylabel('Minimum Number of Conflicts across Population')
plt.savefig('genetic_algorithm_500_gens_vert.png')
plt.show()
```
